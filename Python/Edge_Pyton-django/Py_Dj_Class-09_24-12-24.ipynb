```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Useing arange() inthe place of subplots()
x = np.arange(10, 50, 0.1)
y = np.sin(x)
plt.figure(figsize=(8,4))
plt.plot(x, y, label="Sine X")
plt.title("Sine wave")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#scatter(x, y)
x = input("Enter the value of X: ").split()
y = input("Enter the value of y: ").split()
print(min(x), max(y))
plt.scatter(x, y, marker="*")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.xlim(min(x), max(x))
plt.ylim(min(y), max(y))
plt.title("Scatter")
plt.show()
```

```python
plt.plot([25, 30, 35, 40], [12, 15, 9, 20], label="Line", color="red")
plt.scatter([10, 20, 30, 40], [11, 25, 9, 27], label="Scatter", marker="*", color="green")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.xlim(5, 45)
plt.ylim(5, 30)
plt.title("Scatter")
plt.legend()
plt.show()
```

```python
#Bar
plt.plot([25, 30, 35, 40], [12, 15, 9, 20], label="Line", color="red", linewidth=5)
plt.bar([25, 30, 35, 40], [12, 15, 9, 20], label="Bar", color="green", width=2)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.xlim(20, 45)
plt.ylim(5, 25)
plt.title("Scatter")
plt.legend()
plt.show()
```

```python
#Horizental Bar
plt.plot([25, 30, 35, 40], [12, 15, 9, 20], label="Line", color="red", linewidth=5)
plt.barh([12, 15, 9, 20], [25, 30, 35, 40], label="Bar", color="green", height=2)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.xlim(20, 45)
plt.ylim(5, 25)
plt.title("Scatter")
plt.legend()
plt.show()
```

```python
#Style list
print(plt.style.available)
```

```python
#Use of style
plt.plot([25, 30, 35, 40], [12, 15, 9, 20], label="Line", color="red", linewidth=5)
plt.barh([12, 15, 9, 20], [25, 30, 35, 40], label="Bar", color="green", height=2)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.xlim(20, 45)
plt.ylim(5, 25)
plt.style.use('dark_background')
plt.title("Scatter")
plt.legend()
plt.show()
```

```python
# deffrent color in each bar
color = ['red', 'green', 'blue', 'white']
plt.barh([25, 30, 35, 40], [12, 15, 9, 20], label="Bar", color=color, height=2)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.xlim(5, 25)
plt.ylim(20, 45)
plt.style.use('dark_background')
plt.title("Scatter")
plt.legend()
plt.show()
```

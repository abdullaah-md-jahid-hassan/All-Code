```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Input System 1
# red_x = input("Enter the X Value of RED: ").split()
# red_y = input("Enter the Y Value of RED: ").split()
# blue_x = input("Enter the X Value of BLUE: ").split()
# blue_y = input("Enter the Y Value of BLUE: ").split()

#Input system 2
def input_20_num(c):
    x = []
    for i in range(20):
        x.append(int(input(f"{c}{i+1}=")))
    return x

# Get the red and blue scatter plot values
print("Enter X values for RED:")
red_x = input_20_num("Rx")
print("Enter Y values for RED:")
red_y = input_20_num("Ry")

print("Enter X values for BLUE:")
blue_x = input_20_num("Bx")
print("Enter Y values for BLUE:")
blue_y = input_20_num("By")

plt.scatter(red_x, red_y, label="RED Scatter", marker=".", color="red")
plt.scatter(blue_x, blue_y, label="BLUE Scatter", marker="*", color="blue")

plt.title("Scatter Plot of RED and BLUE")
plt.legend()
plt.show()
```

```python
plt.pie([10, 27], labels=["Done", "Undone"], autopct='%1.2f%%')
plt.title("Pie Chart")
plt.show()
```

```python
fig = plt.figure()
ele_name = ["C", "C++", "Python", "Java", "C#"]
ele_value = [6, 8, 5, 3, 2]
plt.title("PIE chart")
plt.pie(ele_value, labels=ele_name, autopct='%1.2f%%', colors=["red", "green", "blue", "yellow", "white"])
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.axis("equal")
ele_name = ["C", "C++", "Python", "Java", "C#"]
ele_value = [6, 8, 5, 3, 2]
plt.title("PIE chart")
ax.pie(ele_value, labels=ele_name, autopct='%1.2f%%', colors=["red", "green", "#058bf2", "yellow", "#05f244"])
plt.show()
```

```python
np.random.seed(45)
points = 1000
x = np.random.rand(points) #Genarate random value
y = np.random.rand(points)
z = np.random.rand(points)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') #Defining the Size od the graph
ax.scatter(x, y, z, color='red', marker='.')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D graph')
plt.show()
```

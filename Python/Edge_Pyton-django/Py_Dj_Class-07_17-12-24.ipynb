```python
#uerinput
num_point = int(input("Enter the number of the point: "))
value_x = []
value_y = []
for i in range(0, num_point):
    x = int(input(f"Enter the x{i+1}:"))
    y = int(input(f"Enter the y{i+1}:"))
    value_x.append(x)
    value_y.append(y)
fig = plt.figure()
plt.plot (value_x, value_y)
plt.show()
```

```python
#plt.xlim() and plt.ylim() this provide the scaling of x and y axis 
value_x = input("Enter the vlaue of x: ").split()
value_y = input("Enter the vlaue of y: ").split()

fig = plt.figure()
plt.plot (value_x, value_y)
plt.xlim (min(value_x), max(value_x))
plt.ylim (min(value_y), max(value_y))
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.show()
print(fig)
```

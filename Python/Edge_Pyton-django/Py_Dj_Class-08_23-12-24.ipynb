```python
#Library (matplotlib.pyplot)
```

```python
import matplotlib.pyplot as plt
```

```python
#subplots()
```

```python
fig, ax = plt.subplots()
line, = ax.plot([1, 2, 3, 4, 5, 6], [10, 6, 17, 25, 13, 23], label="Line")
line.set_color("red")
line.set_linewidth(5)
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_title("Simple Line")
plt.legend()
plt.show()
```

```python
#Library (numpy)
```

```python
import numpy as np
```

```python
x = np.linspace(10, 50, 400)
y = np.sin(x)
plt.figure(figsize=(8,4))
plt.plot(x, y, label="Sine X")
plt.title("Sine wave")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
all_data = pd.read_csv('10000_student_result.csv') # read CSV
# take Frem to lists
cgpa_list = all_data['CGPA'].tolist()
cgpa_dic = {"A+":0, "A":0, "A-":0, "B+":0, "B":0, "B-":0, "C+":0, "C":0, "D":0, "F":0}
for x in cgpa_list:
    if(x==4): cgpa_dic["A+"]+=1
    elif(x<4 and x>=3.75): cgpa_dic["A"]+=1
    elif(x<3.75 and x>=3.5): cgpa_dic["A-"]+=1
    elif(x<3.5 and x>=3.25): cgpa_dic["B+"]+=1
    elif(x<3.25 and x>=3): cgpa_dic["B"]+=1
    elif(x<3 and x>=2.75): cgpa_dic["B-"]+=1
    elif(x<2.75 and x>=2.5): cgpa_dic["C+"]+=1
    elif(x<2.5 and x>=2.25): cgpa_dic["C"]+=1
    elif(x<2.25 and x>=2): cgpa_dic["D"]+=1
    else: cgpa_dic["F"]+=1
plt.pie(cgpa_dic.values(), labels=cgpa_dic.keys(), autopct='%1.2f%%')
plt.title("Student Result")
plt.show()
```

```python
import tkinter.simpledialog as sd
from tkinter import *
from tkinter import messagebox
from tkinter import ttk
```

```python
#TAking input form user

top = Tk()
top.geometry("500x500")

name = sd.askstring("Input", "Name:")
email = sd.askstring("Input", "Email:")
age = sd.askinteger("Input", "Age:")
price = sd.askinteger("Input", "Price:")
phone = sd.askstring("Input", "Phone:")

name.place(x=50, y=50)
email.place(x=50, y=100)
age.place(x=50, y=150)
price.place(x=50, y=200)
phone.place(x=50, y=250)

top.mainloop()
```

```python
#TAking input form user in a single form (Chat gpt)

# Create the main application window
root = Tk()
root.geometry("300x300")
root.title("Form")

# Function to handle form submission
def submit_form():
    # Retrieve values from the entry fields
    name = name_entry.get()
    email = email_entry.get()
    age = age_entry.get()
    price = price_entry.get()
    phone = phone_entry.get()
    
    # Validate the inputs (optional)
    if not name or not email or not age or not price or not phone:
        messagebox.showwarning("Input Error", "All fields are required!")
        return
    
    # Display entered information (you can modify this to save data elsewhere)
    messagebox.showinfo("Form Submitted", f"Name: {name}\nEmail: {email}\nAge: {age}\nPrice: {price}\nPhone: {phone}")

# Create labels and entry fields
Label(root, text="Name").grid(row=0, column=0, padx=10, pady=5)
name_entry = Entry(root, width=30, bg="lightblue")
name_entry.grid(row=0, column=1, pady=5)

Label(root, text="Email").grid(row=1, column=0, padx=10, pady=5)
email_entry = Entry(root, width=30, bg="lightblue")
email_entry.grid(row=1, column=1, pady=5)

Label(root, text="Age").grid(row=2, column=0, padx=10, pady=5)
age_entry = Entry(root, width=30, bg="lightblue")
age_entry.grid(row=2, column=1, pady=5)

Label(root, text="Price").grid(row=3, column=0, padx=10, pady=5)
price_entry = Entry(root, width=30, bg="lightblue")
price_entry.grid(row=3, column=1, pady=5)

Label(root, text="Phone").grid(row=4, column=0, padx=10, pady=5)
phone_entry = Entry(root, width=30, bg="lightblue")
phone_entry.grid(row=4, column=1, pady=5)

# Add a Submit button
submit_button = Button(root, text="Submit", command=submit_form, bg="lightblue")
submit_button.grid(row=5, column=1, pady=10)

# Run the Tkinter event loop
root.mainloop()
```

```python
#TAking input form user with place(X_position, Y_position)

top = Tk()
top.geometry("500x500")
top.title("Form")

Label(top, text="Name").place(x=50, y=50)
name = Entry(top)
name.place(x=100, y=50)

Label(top, text="Email").place(x=50, y=100)
email = Entry(top)
email.place(x=100, y=100)

Label(top, text="Age").place(x=50, y=150)
age = Entry(top)
age.place(x=100, y=150)

Label(top, text="Price").place(x=50, y=200)
price = Entry(top)
price.place(x=100, y=200)

Label(top, text="Phone").place(x=50, y=250)
phone = Entry(top)
phone.place(x=100, y=250)

top.mainloop()
```

```python
#TAking input form user with grid(row, column, padx, pady, ect)


def submit_form():
    name_g = name.get()
    email_g = email.get()
    age_g = age.get()
    price_g = price.get()
    phone_g = phone.get()
    
    if not name_g.isalpha():
        messagebox.showerror("Validation Error", "Name must contain only letters.")
        return
    if not age_g.isdigit():
        messagebox.showerror("Validation Error", "Age must be a valid integer.")
        return
    if not price_g.isdigit():
        messagebox.showerror("Validation Error", "Price must be a valid integer.")
        return
    if not phone_g.isdigit():
        messagebox.showerror("Validation Error", "Phone must contain only digits.")
        return
    if "@" not in email_g or "." not in email_g:
        messagebox.showerror("Validation Error", "Email must be valid.")
        return
    if not name_g or not email_g or not age_g or not price_g or not phone_g:
        messagebox.showwarning("Input Error", "All fields are required!")
        return
    
    messagebox.showinfo("Form Submitted", f"Name: {name_g}\nEmail: {email_g}\nAge: {age_g}\nPrice: {price_g}\nPhone: {phone_g}")

top = Tk()
top.geometry("500x500")
top.title("Form")

Label(top, text="Name").grid(row=0, column=0, padx=10, pady=10)
name = Entry(top)
name.grid(row=0, column=1, padx=10, pady=10)

Label(top, text="Email").grid(row=1, column=0, padx=10, pady=10)
email = Entry(top)
email.grid(row=1, column=1, padx=10, pady=10)

Label(top, text="Age").grid(row=2, column=0, padx=10, pady=10)
age = Entry(top)
age.grid(row=2, column=1, padx=10, pady=10)

Label(top, text="Price").grid(row=3, column=0, padx=10, pady=10)
price = Entry(top)
price.grid(row=3, column=1, padx=10, pady=10)

Label(top, text="Phone").grid(row=4, column=0, padx=10, pady=10)
phone = Entry(top)
phone.grid(row=4, column=1, padx=10, pady=10)

Button(top, text="Submit", command=submit_form).grid(row=5, column=1, padx=10, pady=10)

top.mainloop()
```
